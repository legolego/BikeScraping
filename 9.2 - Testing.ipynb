```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# Create options for headless mode (optional)
options = Options()
options.headless = True  # Run Chrome in headless mode (without opening a window)

# Manually specify the path to ChromeDriver if it's not in your PATH
driver = webdriver.Chrome(options=options)

# Open a website
driver.get('https://www.pinkbike.com/news/first-ride-race-face-turbine-flat-pedals.html?trk=rss')

# Locate the <div> with class="blog-head" and get the text
blog_head_div = driver.find_element(By.CLASS_NAME, 'blog-head')
blog_head_text = blog_head_div.text

# Print the extracted text
print(blog_head_text)


# Locate the <div> with class="blog-body" and get the text
blog_body_div = driver.find_element(By.CLASS_NAME, 'blog-body')
blog_body_text = blog_body_div.text

# Print the extracted text
print(blog_body_text)

# Close the browser
driver.quit()
```

```python
from bs4 import BeautifulSoup

# Manually specify the path to ChromeDriver if it's not in your PATH
driver = webdriver.Chrome(options=options)

# Open the target RSS feed page
driver.get('https://www.pinkbike.com/pinkbike_xml_feed.php')  # Replace this with the URL of your RSS feed

# Get the page source (the entire HTML of the page)
page_source = driver.page_source

# Parse the HTML/XML using BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find all <link> tags (RSS feed links)
link_tags = soup.find_all('link')
print(soup)

# Extract the href attribute from each <link> tag and print it
for link_tag in link_tags:
    print(link_tag.get('href'))

# Close the driver
driver.quit()
```

```python
# Manually specify the path to ChromeDriver if it's not in your PATH
driver = webdriver.Chrome(options=options)

# Open the target RSS feed page
driver.get('https://www.pinkbike.com/pinkbike_xml_feed.php')  # Replace with your actual RSS feed URL

# Use JavaScript to get the raw XML source
xml_content = driver.execute_script("return document.documentElement.outerHTML;")

# Parse the XML using BeautifulSoup
soup = BeautifulSoup(xml_content, 'xml')

# Find all <link> tags
link_tags = soup.find_all('link')

# Extract the href attribute from each <link> tag and print it
for link_tag in link_tags:
    print(link_tag.get('href'))

# Close the driver
driver.quit()
```
